```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

# Make the graphs a bit prettier, and bigger
plt.style.use('ggplot')

# This is necessary to show lots of columns in pandas 0.12. 
# Not necessary in pandas 0.13.
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)

plt.rcParams['figure.figsize'] = (15, 5)
plt.rcParams['font.family'] = 'sans-serif'
```

```python
complaints = pd.read_csv('https://raw.githubusercontent.com/jvns/pandas-cookbook/master/data/311-service-requests.csv',dtype='unicode')
```

```python
complaints[:3]
```

```python
ComplaintType = complaints['Complaint Type'].value_counts()
ComplaintType[:20].plot(kind='bar')

NoiseComplaints = complaints[complaints['Complaint Type'] == "Noise - Street/Sidewalk"]
NoiseComplaints[:5]
```

```python
Boroughs = complaints['Park Borough'].value_counts()
Boroughs[:20]
```

```python
CountofNC = NoiseComplaints['Park Borough'].value_counts()
CountofNC.plot(kind='bar')
```

```python
TotalComplaints = complaints['Park Borough'].value_counts()
PercofNC = (CountofNC / TotalComplaints)*100
PercofNC.plot(kind='pie')
```
